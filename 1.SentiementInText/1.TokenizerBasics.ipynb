# Tokenizer Basics:

In most of NLP tasks, the intial step in prepraring your data is to extract a vocabulary of words from your *corpus*(i.e input text). You will need to define how to represent the texts into numerical representations which can be used to train a neural network. These representations are called **token** and Tensorflow and keras makes it easy to generate these using its APIs. You will see how to do that in the next cells.
